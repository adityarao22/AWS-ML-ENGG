```python
class User1:
    def __init__(self,pen,pencil):
        print("Constructor of User1")
        self.pen=pen
        self.pencil=pencil
    def penGet(self):
        print("pen is working :",self.pen)
    def pencilGet(self):
        print("pencil is working :",self.pencil)

class User2(User1):
    def __init__(self,pen,pencil,erasor):
        print("Constructer of User2")
        self.pen=pen
        self.pencil=pencil
        self.erasor=erasor

#object
u=User2("blue","black","doms")
u.penGet()
```

```python
class User1:
    def __init__(self,pen,pencil):
        print("Constructor of User1")
        self.pen=pen
        self.pencil=pencil
    def penGet(self):
        print("pen is working :",self.pen)
    def pencilGet(self):
        print("pencil is working :",self.pencil)

class User2(User1):
    def __init__(self,pen,pencil,erasor):
        print("Constructer of User2")
       # self.pen=pen
       # self.pencil=pencil
        super().__init__(pen,pencil)
        self.erasor=erasor

#object
u=User2("blue","black","doms")
u.penGet()
```

```python
class Aditya:
    def __init__(self,bike):
        self.bike=bike
    def usebike(self):
        print("bike used by :",self.bike)

class Blaze(Aditya):
    def __init__(self,name,bike):
        self.name=name
        #approach 1
        #self.bike=bike
        #approach 2
        super().__init__(bike)
    def usename(self):
        print("name is :",self.name)
u=Blaze("Blaze","RF by Blaze")
u.usename()
u.usebike()
```

```python
#encapsulation
class Student:
    def __init__(self,name,roll_no): 
        self.__data=name         #name is attribute
        self.roll_no=roll_no  #roll_no is attribute
    def getName(self,password=None):
        if password == "123456":
           print(self.__data)
        else:
            print("password not match")
s= Student("Aditya",13)
#print(s.__data)   #no attribute
print(s.roll_no)
s.getName("123456")
```

```python
#example
class Student:
    def __init__(self,name,marks):
        self.name=name
        self.__marks=marks
    def getMarks(self):
        print(self.__marks)
    def updateMarks(self,marks):
        self.__marks=marks
        print("Updated marks :",self.__marks)

s=Student("Aditya",95)
print(s.name)
s.getMarks()
s.updateMarks(100)
```
