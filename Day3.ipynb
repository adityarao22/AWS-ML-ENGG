```python
marks=int(input("Enter marks: "))
if marks>=95:
    print("Grade A")
elif marks>=90:
    print("Grade B")
elif marks>=75:
    print("Grade C")
elif marks>=60:
    print("Grade D")
else:
    print("Fail")
```

LIST


```python
l=[10,20,30]
#positive indexing
print(l[0],l[1],l[2]) #print element from last to first
#negative indexing
print(l[-1],l[-2],l[-3]) #print element from first to last

#mutable
l[0]=100
l[-1]=300
print(l)
```

Dictionary

```python
#obj = {key:value}
#Example obj={"Name":"Aditya rao"}

d={"Name": "Aditya rao"}
print(d)
print(d["Name"])

#example
d={"Full name": "Aditya rao"}
print(d)
print(d["Full name"])
```

```python
d={"age":20, "name": "Aditya rao"}
print(d["age"])
print(d["name"])
```

```python
d={"users": ["Aditya rao","Blaze","Golu"]}
print(d["users"])
print(d["users"][0])
print(d["users"][1])
```

```python
d={"users": [{"name":"Hello"},{"name":"Arya"}]}
print(d["users"][0]["name"])
print(d["users"][1]["name"])
```
