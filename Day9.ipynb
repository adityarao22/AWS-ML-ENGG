```python
# array opr on zeros | ones | full

import numpy as np

arr = np.zeros(5)

print(arr)
print(arr.shape)   # (5,)
print(arr.ndim)    # 1
print(arr.size)    # 5
print(arr.dtype)   # float

# 2d
print("2d zeros start")

arr1 = np.zeros((3, 2))

print(arr1)
print(arr1.shape)   # (3, 2)
print(arr1.ndim)    # 2
print(arr1.size)    # 6
print(arr1.dtype)   # float
```

```python
# array opr on zeros | ones | full

import numpy as np

arr = np.ones(5)

print(arr)
print(arr.shape)   # (5,)
print(arr.ndim)    # 1
print(arr.size)    # 5
print(arr.dtype)   # float

# 2d
print("2d ones start")

arr1 = np.ones((3, 2))

print(arr1)
print(arr1.shape)   # (3, 2)
print(arr1.ndim)    # 2
print(arr1.size)    # 6
print(arr1.dtype)   # float
```

Array Reshaping

```python
#1D array

import numpy as np
arr=np.arange(6)
print(arr)

#2D array
import numpy as np
arr1=np.arange(1,13).reshape(3,4)
print(arr1)

#another approach
arr2=arr.reshape(2,3) #2rowa and 3 columns
print(arr2)
```

```python
#flatten (convert to 1D)

import numpy as np
#2D
arr=np.arange(6)
arr1=arr.reshape(2,3)
print(arr1)

#flatten
arr_new=arr1.flatten()
arr_new[0]=100
print(arr1)
```

```python
#ravel(convert to 1D)

import numpy as np 
#2D
arr=np.arange(6).reshape(2,3)
print(arr)

#ravel
arr_new1=arr.ravel()
arr_new1[0]=200
print(arr)
```

flatten --> it can create a copy of original 
Ravel --> It can return view(original data)

```python
#transpose

import numpy as np
arr=np.arange(6).reshape(2,3)
print(arr)

print(np.transpose(arr))
```
